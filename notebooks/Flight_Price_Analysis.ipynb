```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
print("Numeric columns:", numeric_cols)
```

```python
print("Shape before outlier removal:", df.shape)
plt.boxplot(df[numeric_cols])
plt.title('Boxplot of Numeric Columns')
plt.show()
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower_bound, upper_bound)
```

```python
print("-"*50)
plt.figure(figsize=(10, 8))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
print("-"*50)
print("Shape after outlier removal:", df.shape)
plt.boxplot(df[numeric_cols])
plt.title('Boxplot of Numeric Columns After Outlier Removal')
plt.show()
print("-"*50)
print("Dataset after outlier removal:")
print(df.head())
```

# Price Distribution
Check how ticket prices are spread

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("price Distribution across Tickets")
plt.xlabel("price (INR)")
plt.ylabel("Count")
plt.show()
```

# Ticket Count per Airline

```python
plt.figure(figsize=(8,5))
sns.countplot(y='airline', data=df, order=df['airline'].value_counts().index)
plt.title("Number of Flights per airline")
plt.xlabel("Count")
plt.ylabel("airline")
plt.show()
```

# Price vs Airline
Understand price variation by airline:

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='airline', y='price', data=df, showfliers=False)
plt.title("Price Distribution by Airline")
plt.xticks(rotation=45)
plt.show()
```

# Stops vs Price
Compare price trends by number of stops:

```python
sns.boxplot(x='stops', y='price', data=df, showfliers=False)
plt.title("Price vs Number of Stops")
plt.show()
```

#  Days Left vs Price
See if booking early affects price:

```python
sns.scatterplot(x='days_left', y='price', data=df.sample(5000), alpha=0.3)
plt.title("Price vs Days Left Until Trip")
plt.xlabel("days_left")
plt.ylabel("price (INR)")
plt.show()
```

#  Duration vs Price

```python
sns.jointplot(x='duration', y='price', data=df.sample(5000), kind='scatter', height=7, alpha=0.3)
plt.suptitle("Flight Duration vs price", y=1.02)
plt.show()
```

# Class-Based Comparison (Economy vs Business)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='class', y='price', data=df, showfliers=False)
plt.title("Price: Economy vs Business Class")
plt.show()
```

# pivot average price between route combinations

```python
pivot = df.pivot_table(index='source_city', columns='destination_city', values='price', aggfunc='mean')
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Price by Route (Source → Destination)")
plt.xlabel("Destination")
plt.ylabel("Source")
plt.show()
```
